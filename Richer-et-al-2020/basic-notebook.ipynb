## Reuse Functions That You Create In Scripts
https://www.earthdatascience.org/courses/earth-analytics/multispectral-remote-sensing-data/source-function-in-R/

```python
%%R 
source("scripts/functions.R")
```

```python
%%R 
listFunctions <- function(filename) {
  temp.env <- new.env()
  sys.source(filename, envir = temp.env)
  functions <- lsf.str(envir=temp.env)
  rm(temp.env)
  return(functions)
}
```

```python
%%R 
listFunctions('scripts/functions.R')
```
